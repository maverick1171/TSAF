```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def calculate_moving_average():
    """
    Demonstrates fitting a trend using the Moving Average Method.
    We will calculate both 3-Year and 5-Year moving averages.
    """
    print("--- Fitting Trend by Moving Average Method ---")

    # Creating a sample dataset (e.g., yearly production or sales)
    # This data has a general upward trend but fluctuates year-to-year.
    data = {
        'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
        'Sales': [45, 48, 42, 56, 52, 60, 58, 65, 63, 70, 68]
    }
    
    # Load data into a Pandas DataFrame
    df = pd.DataFrame(data)

    # Calculate 3-Year Moving Average
    # center=True places the average exactly in the middle year of the window (e.g., year 2 for 1,2,3)
    df['3-Year MA'] = df['Sales'].rolling(window=3, center=True).mean()

    # Calculate 5-Year Moving Average
    df['5-Year MA'] = df['Sales'].rolling(window=5, center=True).mean()

    # Round the values to 2 decimal places for clean exam output
    df['3-Year MA'] = df['3-Year MA'].round(2)
    df['5-Year MA'] = df['5-Year MA'].round(2)

    print("\nCalculated Moving Average Table:")
    # Fill NaN (Not a Number) values with a placeholder like '-' for better display
    print(df.fillna('-').to_string(index=False))

    # Create the visualization to show the 'smoothing' effect
    plt.figure(figsize=(10, 6))
    
    # Plot original data (dashed line with markers to show fluctuations)
    plt.plot(df['Year'], df['Sales'], marker='o', linestyle='--', color='gray', label='Original Sales Data')
    
    # Plot 3-Year Trend (Solid Blue Line)
    plt.plot(df['Year'], df['3-Year MA'], marker='s', linewidth=2, color='blue', label='3-Year Moving Average')
    
    # Plot 5-Year Trend (Solid Red Line)
    plt.plot(df['Year'], df['5-Year MA'], marker='^', linewidth=2, color='red', label='5-Year Moving Average')

    # Add titles, labels, and grid for a professional look
    plt.title('Fitting of Trend by Moving Average Method', fontsize=14, fontweight='bold')
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Sales Volume', fontsize=12)
    
    # Ensure all years are shown on the X-axis
    plt.xticks(df['Year']) 
    
    plt.legend(loc='best')
    plt.grid(True, linestyle=':', alpha=0.7)
    
    # Display the plot
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    calculate_moving_average()
```

### <i> What is Moving Average Method ?
### "It removes the short-term peaks and troughs to isolate the long-term trend".
### <i> How does it works ?
### "It does this by averaging data points over a specific sliding window of time, which smooths out the sudden fluctuations."
